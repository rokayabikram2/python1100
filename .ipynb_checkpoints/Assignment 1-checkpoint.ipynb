# WAP to swap values between two variable

```python
##

num1 = 2.5
num2 = 0.36


print("num1 =",num2)
print("num2 =",num1)
```

```python
##
num1 = 2.5
num2 = 0.36

print(f"num1 = {num2}")
print(f"num2 ={num1}")
```

```python
num1 = 2.5
num2 = 0.36

temp =num1
num1 =num2
num2 =temp

print(f"value of num1 after swaping = {num1}")
print(f"value of num2 after swaping = {num2}")
```

```python
num1 = 2.5
num2 = 0.36

num =num1
num1 =num2
num2 =num

print(f"value of num1 after swaping = {num1}")
print(f"value of num2 after swaping = {num2}")
```

```python
num1 = 2.5
num2 = 0.36

num1 , num2 =num2 , num1
```

# Avg speed = miles/min


```python
miles = 10/1.609
minutes = 42 + 42/60
average = miles/minutes
print(f"Average speed ={average} miles/min")
```

# Avg speed = miles/sec

```python
miles = 10/1.609
sec= 42*60 + 42
average = miles/sec
print(f"Average speed in miles/sec ={average}")
```

# WAP to find out the volume of radius of sphere having radius 5

```python
PI = 3.14
r = 5
volume =(4/3)*PI*r**3
print(f"Volume of Sphere = {volume}")
```
